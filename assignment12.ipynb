# defining a function for regularization and making a optimal set of model parameter with regularization

```python
def regularizing(x_tilda,ydata,nlambda,p,arr):
    x_tilda_size = x_tilda.size
    reshape1 = (int)(x_tilda_size * p)
    reshape2 = (int)(x_tilda_size / p)
    mat_b = nlambda * np.eye(p)
    kroneck_left = np.kron(x_tilda,mat_b)
    kroneck_right = np.kron(ydata,0)
    
    theta = x_tilda.transpose() * y2 / np.linalg.norm(x_tilda)
    left_reshape = kroneck_left.reshape(reshape1,p)
    left = np.dot(left_reshape,theta)
    left_reshape = kroneck_left.reshape(reshape1,p)
    
    left_calc = left - kroneck_right
    
    right_calc = nlambda * np.linalg.norm(pow(theta,2))
    
    epsil = left_calc + right_calc
    epsil = epsil.reshape(reshape2,reshape1)
    minimum_val = np.zeros(reshape2,)
    for i in range(reshape2):
        arr[i] = epsil[i].min()
```

```python
minimum_val = np.zeros(1001,)
minimum_val2 = np.zeros(1001,)
minimum_val3 = np.zeros(1001,)
minimum_val4 = np.zeros(1001,)
minimum_val5 = np.zeros(1001,)

regularizing(x_tilda,y2,pow(2,1),p,minimum_val)
regularizing(x_tilda,y2,pow(2,2),p,minimum_val2)
regularizing(x_tilda,y2,pow(2,3),p,minimum_val3)
regularizing(x_tilda,y2,pow(2,4),p,minimum_val3)
regularizing(x_tilda,y2,pow(2,6),p,minimum_val3)
```
