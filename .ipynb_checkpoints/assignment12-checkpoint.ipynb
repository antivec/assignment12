# import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num     = 1001
std     = 5 
```

```python
def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f
```

# defining a function for calculating model parameter

```python
def calc_model_para(x, y):
    x_trans = x.transpose()
    kk = np.linalg.inv(np.dot(x_trans,x))
    kkl = np.dot(kk,x_trans)
    model = np.dot(kkl,y2)
    return model
```

# x  : x-coordinate data
# y1 : (clean) y-coordinate data 
# y2 : (noisy) y-coordinate data

```python
n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x) 		# clean points
y2      = y1 + nn * std		# noisy points
y2
```

# plot noisy datas and clean datas

```python
plt.plot(x, y1, 'b.',label = "clean datas") 
plt.plot(x, y2, 'k.', label = "noisy datas")
plt.legend(loc='best')
plt.show()
```

# plot polynomial curves that fit the noisy data with 
# p = 6 ~ 15

```python
for p in range(6,16):
    x_tilda = np.array([x, np.ones(len(x))]).transpose()
    for degree in range(2, p):
        x_tilda = np.append(np.array(np.power(x, degree))[np.newaxis].transpose(), x_tilda, axis=1)
        model_func = calc_model_para(x_tilda, y2)
        dot = np.dot(x_tilda,model_func)
    plt.plot(x,dot,'k.',label = p)
    plt.legend(loc='upper left')
    plt.show()
```

# with fixed p = 6, start calculating 	
\begin{equation*}
\ {\varepsilon}(\theta ;\lambda) = \|A\theta - y\|^2_2 + \lambda\|\theta\|^2_2
\end{equation*}

```python
p = 6
x_tilda = np.array([x, np.ones(len(x))]).transpose()
for degree in range(2, p):
    x_tilda = np.append(np.array(np.power(x, degree))[np.newaxis].transpose(), x_tilda, axis=1)

x_norm=np.linalg.norm(x_tilda)
theta = np.dot(x_tilda.transpose(),y2) / x_norm
kkkk = np.dot(x_tilda,theta) - y2
```

# regularize and vectorize
\begin{equation*}
\ min\|\begin{bmatrix}
A
\\ \lambda I
\end{bmatrix}\theta - \begin{bmatrix}
y
\\ 0
\end{bmatrix} \|^2 + \lambda \|\theta_0^2 + \theta_1^2 + .... + \theta_p^2\|^2
\end{equation*}

\begin{equation*}
\lambda = 2^2
\end{equation*}

# use kronecker product to vectorize
\begin{equation*}
left= A\otimes \lambda I , right =y \otimes 0
\end{equation*}

```python
lambd = pow(2,2)
mat_b = lambd * np.eye(6)

kroneck_left = np.kron(x_tilda,mat_b)
kroneck_right = np.kron(y2,0)
```

\begin{equation*}
\theta = \frac{A^Ty_2}{\|A\|^2}
\end{equation*}

\begin{equation*}
\ leftcalc =  \|\begin{bmatrix}
A
\\ \lambda I
\end{bmatrix}\theta - \begin{bmatrix}
y
\\ 0
\end{bmatrix} \|^2
\end{equation*}

```python
theta = x_tilda.transpose() * y2 / np.linalg.norm(x_tilda)
left_reshape = kroneck_left.reshape(36036,6)
left = np.dot(left_reshape,theta)
left_reshape = kroneck_left.reshape(36036,6)
```

```python
left_calc = left - kroneck_right
```

\begin{equation*}
\ rightcalc = \lambda \|\theta_0^2 + \theta_1^2 + .... + \theta_p^2\|^2
\end{equation*}

```python
right_calc = lambd * np.linalg.norm(pow(theta,2))
```

\begin{equation*}
\ {\varepsilon}(\theta ;\lambda) = leftcacl + righcalc
\end{equation*}

```python
epsil = left_calc + right_calc
epsil = epsil.reshape(1001,36036)
minimum_val = np.zeros(1001,)
for i in range(1001):
    minimum_val[i] = epsil[i].min()
```

```python
plt.plot(x,minimum_val,'b.')
plt.show()
```

```python
def regularizing(x_tilda,ydata,nlambda,p,arr):
    x_tilda_size = x_tilda.size
    reshape1 = (int)(x_tilda_size * p)
    reshape2 = (int)(x_tilda_size / p)
    mat_b = nlambda * np.eye(p)
    kroneck_left = np.kron(x_tilda,mat_b)
    kroneck_right = np.kron(ydata,0)
    
    theta = x_tilda.transpose() * y2 / np.linalg.norm(x_tilda)
    left_reshape = kroneck_left.reshape(reshape1,p)
    left = np.dot(left_reshape,theta)
    left_reshape = kroneck_left.reshape(reshape1,p)
    
    left_calc = left - kroneck_right
    
    right_calc = nlambda * np.linalg.norm(pow(theta,2))
    
    epsil = left_calc + right_calc
    epsil = epsil.reshape(reshape2,reshape1)
    minimum_val = np.zeros(reshape2,)
    for i in range(reshape2):
        arr[i] = epsil[i].min()
```

```python
minimum_val2 = np.zeros(1001,)
regularizing(x_tilda,y2,pow(2,10),p,minimum_val2)
```
